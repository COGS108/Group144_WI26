## Research Question

Which academic and experiential factors (e.g., GPA, internship experience, project experience) are most predictive of internship outcomes in the technology industry?

## Background and Prior Work

Since freshman year, we've been given lots of advice on the multiple factors that go into getting a job post-graduation. However, some mentors give mixed advice, and we want to pinpoint exactly what combination of GPA, professional experience, and extracurriculars makes for the highest likelihood of employment post-graduation. And expanding on this question, we want to tackle this across multiple fields and see whether different students should prioritize certain variables for their own success. After light research, we found many sources that took into account variables such as GPA, internship experience, project experience, technical skills, and soft skills, such as behavioral interviews and networking. While many of these factors may be difficult to quantify in our project, they give us a good starting point for where we should look.

One of the first papers we've looked into was from Bowling Green State University<sup><a href="#fn1" id="ref1">1</a></sup>, which largely studied undergraduate business students for the correlation between academic and co-curricular factors and full-time job placement post graduation and starting salaries. Some of their findings were that internship experience, GPA, and area of study were the most accurate predictors for full-time employment after graduation. Many of these same factors had a high correlation with starting salary; however, surprisingly, participation in college organizations showed some correlation, but to a weaker extent. Another paper from the Journal of Vocational Behavior<sup><a href="#fn2" id="ref2">2</a></sup> studied the effectiveness of internship experiences on work placement post graduation. While this paper had some interesting findings questioning the effectiveness of internships, they lacked a data-driven approach that would be of much use to our project. Going forward, it would be beneficial to look for research that studies the efficacy of academic and co-curricular factors in multiple fields so we can draw conclusions for multiple industries.

Additionally, others examined whether internships directly translate to employment and job advantages. Nunley et al. (2016)<sup><a href="#fn3" id="ref3">3</a></sup> analyze the outcomes of students obtaining internships early on in their career, which is associated with a higher likelihood of receiving a job offer after graduation. This is especially the case when the internship is relevant to the field the student is going into or studying. This demonstrates that internships are a signal to employers of having readily equipped skills that their education alone does not prepare them for. However, this article does not account for how various factors interact across various industries and which of these factors matter more for hiring purposes. 

Although prior research highlights the importance of academic and experiential factors, much of the previous research done on this field focuses on a narrow population or focuses on one sole predictor on employability. There is limited work on the importance of applying employability to a larger scale as well as using multiple predictors across various industries using a Data Science approach. Our project aims to address this gap by examining multiple futures like GPA, internship experience, skills, and other factors jointly correlate to employment outcomes. We hope that this project will give insights to students looking for work experience and to guide them in what to focus on in their professional development.

### Major Concerns
There is little publicly available documentation about where and how this dataset was collected on Kaggle, who the students are, or what institutions they come from, which makes it difficult to assess representativeness. Without the detailed metadata or collection context, it could bias results if the sample is not representative of the broader undergraduate population. 

Given the fact that this dataset was contributed by an individual user on Kaggle, if the dataset originates from a specific educational system or type of institution, institution-specific patterns (grading, internship access mechanisms) could drive associations rather than true predictive relationships relevant to the broader tech internship ecosystem, which reduce its power to address our RQ. 

Because the dataset documentation does not explicitly describe missing values or preprocessing, there’s a risk of listwise deletion or incomplete cases biasing the analysis if missingness is non-random (e.g., students with worse outcomes might be less likely to report certain attributes).

This dataset on Kaggle is not verified, which weakens the credibility and makes it difficult to evaluate construct validity or sampling integrity. There is probability for data fabrication, which reduces the predictive power for the model based on these data.

The dataset is already clean on Kaggle, so we suspect some other people might have already worked on this data, which might introduce distortion to the data.

```python
def prob_impute(df):
    df = df.copy()
    for col in df.columns:
        s = df[col].copy()
        num_null = s.isna().sum()
        fill_vals = np.random.choice(s.dropna(), num_null)
        s[s.isna()] = fill_vals
        df[col] = s
    return df
dropped = prob_impute(dropped)
```
